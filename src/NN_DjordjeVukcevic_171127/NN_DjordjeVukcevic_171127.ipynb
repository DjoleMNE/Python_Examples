```python
def gen_samples(w, r, d, num):
    samples = []
    for i in range(num):
        l = r + np.random.rand() * w
        t = 2 * np.pi * np.random.rand()
        x0 = l * np.cos(t)
        x1 = l * np.sin(t)
        y = 1
        if t > np.pi:
            y = -1
            x0 += r + w/2
            x1 -= d
        samples.append((np.array([x0, x1]), np.array([y])))
    return np.array(samples)

def gen_samples2(l, h, num, function):
    samples = []
    for i in range(num):
        x = np.random.uniform(low = l, high = h)
        y = function(x)
        samples.append((np.array([x]), np.array([y])))
    return np.array(samples)

def kmeans(K, samples, dimension = 2):
    assign = np.zeros(len(samples))
    choice = np.random.randint(len(samples), size=K)
    means = [samples[c][0] for c in choice]
    variance = np.zeros(K)
    changed = True
    while changed:
        old = np.copy(assign)
        for i in range(len(samples)):
            if(dimension == 1):
                dists = np.abs(samples[i][0] - means) 
            else:
                dists = np.linalg.norm(samples[i][0] - means, axis=1) ** 2
            assign[i] = np.argmin(dists)
        changed = np.any(assign != old)
        for k in range(K):
            sel = np.where(assign == k)
            means[k] = np.mean(samples[sel,0])
            incluster = samples[sel,0][0]
            dists = [np.linalg.norm(a - means[k]) for a in incluster]
            variance[k] = np.var(dists)
            
    return (assign.astype(int), means, variance)
```

```python
K = 6
r = 8
data = {}
for t in range(r):
    d = 1 - t
    samples = gen_samples(1, 7, d, 1000)
    (clusters, means, variance) = kmeans(K, samples)
    data[t] = (d, samples, clusters, means, variance)
```

```python
colors = ['red', 'blue', 'pink', 'green', 'brown', 'purple']

fig, ax = plt.subplots(4, 2,figsize=(4*8,4*8),sharex=True,sharey=True)
ax = ax.flatten()
for t in range(r):
    (d, samples, clusters, means, variance) = data[t]
    ax[t].axis('equal')
    ax[t].grid()
    ax[t].set_title('d = {:d}:'.format(d))
    ax[t].set_xlabel('x')
    ax[t].set_ylabel('y')
    for i in range(len(samples)):
        s = samples[i]
        cluster = clusters[i]
        x = s[0]
        c = colors[cluster]
        ax[t].scatter(x[0], x[1],c=c, marker='o')
    for i in range(K):
        mean = means[i]
        c = colors[i]
        ax[t].scatter(mean[0], mean[1], c='black', marker='D', s=100)

    #ax[t].title = 'test'
fig.suptitle('Kmeans clusters over d', fontsize=18)
plt.show(fig)
```

```python
class RBFNet(object):
    def __init__(self, k_num, epochs, dimension = 2):
        self.k = k_num
        self.dimension = dimension
        self.epochs = epochs
        self.w = np.atleast_2d(np.random.random(self.k)).T
    
    def rbf(self, x, c, s):
        if (self.dimension == 1):
            result = np.exp(-1 / (2 * s) * (x-c)**2)
        else:
            result = np.exp(-1 / (2 * s) * np.linalg.norm(x-c)**2)
        return result
    
    def activation(self, X):
        a = np.zeros((len(X), self.k))
        for i in range(len(X)):
            for j in range(self.k):
                a[i,j] = self.rbf(X[i], self.center[j], self.variance[j])
        return a
    
    def train_lms(self, samples):
        self.clusters, self.center, self.variance = kmeans(self.k, samples, self.dimension)
        a = self.activation(samples[:,0])
        a_inv = np.linalg.pinv(a)
        self.w = np.dot(a_inv,samples[:,1])
    
    def train_rls(self, samples, epochs, start_lr, end_lr):
        self.clusters, self.center, self.variance = kmeans(self.k, samples, self.dimension)
        maxdist = 0
        for c0 in self.center:
            for c1 in self.center:
                dist = np.linalg.norm(c0-c1)
                if dist > maxdist:
                    maxdist = dist
        self.variance[:] = maxdist / np.sqrt(2 * self.k)
        self.w = np.atleast_2d(np.zeros(self.k)).T
        lr_space = np.linspace(start_lr, end_lr, epochs)
        P = np.eye(len(samples[:,0])) * 0.01 ** -1
        act = self.activation(samples[:,0])
        top = np.dot(P, np.dot(act, np.dot(act.T, P)))
        bottom = (1 + np.dot(act.T, np.dot(P, act)))
        P = P - top / bottom
        print(P)
        print(P.shape)
        for epoch, lr in zip(range(epochs), lr_space):
            pass
            #print(epoch)
            #print(lr)

    def _lim(self, y):
        if y > 0:
            return 1
        else:
            return -1
        
    def predict(self, samples):
        a = self.activation(samples)
        Y = np.dot(a, self.w)
        return [self._lim(y[0]) for y in Y]
    
    def predict2(self, samples):
        a = self.activation(samples[:,0])
        Y = np.dot(a, self.w)
        return Y
```

```python
K = 20
ds = [-5, -6]
fig, ax = plt.subplots(1, 2,figsize=(8 * len(ds), 6),sharex=True,sharey=True)
ax = ax.flatten()

x_min, x_max = -10, 15
y_min, y_max = -5, 10
for i in range(len(ds)):
    d = ds[i]
    train_set = gen_samples(1, 6, d, 1000)
    test_set = gen_samples(1, 6, d, 2000)
    rb = RBFNet(K, 1)
    rb.train_lms(train_set)
    Y = rb.predict(test_set[:,0])
    ax[i].axis('equal')
    ax[i].grid()
    ax[i].set_title('d = {:d}:'.format(d))
    for x, y in zip(test_set[:,0], Y):
        c = 'r'
        if y < 0:
            c = 'b'
        ax[i].scatter(x[0], x[1], c=c)
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.25), np.arange(y_min, y_max, 0.25))
    grid_samples = np.c_[xx.ravel(), yy.ravel()]
    z = rb.predict(grid_samples)
    z = np.array(z).reshape(xx.shape)
    ax[i].contourf(xx, yy, z, alpha=0.4)
plt.show(fig)
```
